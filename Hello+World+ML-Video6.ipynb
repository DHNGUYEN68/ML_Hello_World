```python
# Export decision tree in PDF
#Video # 2 (cont.)
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
# Three rows for testing data
# The rest for training data
# Never use testing data for training the model
# Indices for test data 
test_idx = [0,5,100]

#Training data
train_target = np.delete(iris.target, test_idx)
train_data = np.delete(iris.data, test_idx, axis=0)

#Testing data
test_target = iris.target[test_idx]
test_data = iris.data[test_idx]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(train_data, train_target)

print(iris.feature_names)
print (test_data)
print (test_target)
print (clf.predict(test_data))
res_predict = clf.predict(test_data)
# Use dictionary to have meaningful labels
flower={0:'setosa',1:'versicolor',2:'virginica'}
type(res_predict)
for i in range(len(res_predict)):
    print ('Current flower:',i,"->" ,flower[res_predict[i]])

# ************** Visualize the decision tree ***************
# To install the required packages
#pip3 install pydot
#pip3 install graphviz
#conda install python-graphviz
#conda install graphviz
#************** Visualize the decision tree *****************
from sklearn.externals.six import StringIO
import pydot
import graphviz 
dot_data = StringIO()
dot_data=tree.export_graphviz(clf, out_file=None, 
                         feature_names=iris.feature_names,  
                         class_names=iris.target_names,  
                         filled=True, rounded=True,  
                         impurity=False)  

# Export in pdf
graph = graphviz.Source(dot_data) 
# View = True, opens in Pdf reader
graph.render("iris", view=True) 
```

```python
# Video #4
# This cell shows how to check the accuracy of predictions (versus test data)
from sklearn.datasets import load_iris
iris = load_iris()
# X is a Matrix (capital lettre)
# y is a vector (lower case)
X=iris.data
y=iris.target

from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
# Split fithty fithty to keep 50% of data to use as test data 

from sklearn import tree

# Use decision tree classifier un scikit learn package
my_classifier = tree.DecisionTreeClassifier()
my_classifier.fit(X_train,y_train)

# Make predictions on the test data which are not used as training data !
predictions = my_classifier.predict(X_test)

print (predictions)

# Now we assess the accuracy of predictions : the rate should be superior 90%
# http://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html
# Proto : sklearn.metrics.accuracy_score(y_true, y_pred, normalize=True, sample_weight=None)
# Returns the rate of success : normalize=True
# Returns the number of correctly classified samples: normalize=False
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions, normalize=True))
```

```python
#Now we enhance the classifier logic with KNN algorithm (which compares distance)
#determines the distance between two datapoints
#import random
#class ScrappyKNN():
#    def fit(self,X_train,y_train):
#        self.X_train = X_train
#        self.y_train = y_train
#    def predict(self,X_test):
#        predictions = []
#        for row in X_test:
#            label = random.choice(self.y_train)
#            predictions.append(label)
#        return(predictions)
from sklearn.datasets import load_iris
iris = load_iris()
# X is a Matrix (capital lettre)
# y is a vector (lower case)
X=iris.data
y=iris.target 

from scipy.spatial import distance
def euc(a,b): 
    return distance.euclidean(a,b)

#defines customized classifier (based on KNN)
#
class S_KNN(): 
    def fit(self, X_train,y_train):
        self.X_train = X_train
        self.y_train = y_train
        
    def predict(self,X_test): 
        predictions=[] 
        for row in X_test: 
            label = self.closest(row) 
            predictions.append(label) 
        return predictions 
    def closest(self,row): 
        best_distance = euc(row, self.X_train[0]) 
        best_index=0
#The closest is based on the shortest distance between two datapoints
#we use the euclidian distance from scipy library
        for i in range(1,len(self.X_train)):
            dist = euc(row,self.X_train[i])
            if dist < best_distance:
                best_dist = dist
                best_index = i
        return(self.y_train[best_index])
#We use the same pipeline, but another classifier (ScrappyKNN built by own)
#
#This cell shows how to check the accuracy of predictions (versus test data)
#
#from sklearn.datasets import load_iris iris = load_iris()
#
#X is a Matrix (capital lettre)
#
#y is a vector (lower case)
#
#X=iris.data y=iris.target

from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5)
3
#Split fithty fithty to keep 50% of data to use as test data
#Use KNN classifier un scikit learn package
my_classifier = S_KNN() 
my_classifier.fit(X_train,y_train)
#
#Make predictions on the test data which are not used as training data !
#
predictions = my_classifier.predict(X_test)

print(X_test.size)
print(X_train.size)
print(X.size)
#Check the size of prediction
print ("Len of predictions list", len(predictions))

# Now we assess the accuracy of predictions : the rate should be superior 90%
# http://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html
# Proto : sklearn.metrics.accuracy_score(y_true, y_pred, normalize=True, sample_weight=None)
# Returns the rate of success : normalize=True
# Returns the number of correctly classified samples: normalize=False
from sklearn.metrics import accuracy_score
print("Score:", accuracy_score(y_test,predictions, normalize=True))
# Check the size of testing labels data
print("Size of testing labe ndarray", y_test.size)
```
